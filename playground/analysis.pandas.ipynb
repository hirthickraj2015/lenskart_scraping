```python
# Create new columns with the first 3 characters of the 'pincode' column
customers_df['pincode_3'] = customers_df['pincode'].str.slice(0, 3)
stores_df['pincode_3'] = stores_df['pincode'].str.slice(0, 3)

# Perform the merge
merged_df = pd.merge(customers_df, stores_df, left_on='pincode_3', how='inner')
len(merged_df)
```

```python
def assign_state(pincode):
    pincode = int(pincode[:3])  # Get the first 3 digits of the pincode
    for state, (start, end) in state_pincode_ranges.items():
        if start <= pincode <= end:
            return state
    return 'Unknown'  # Return 'Unknown' if the pincode does not match any state

# Assign a state to each customer
customers_df['state'] = customers_df['pincode'].apply(assign_state)

display(customers_df)
```
